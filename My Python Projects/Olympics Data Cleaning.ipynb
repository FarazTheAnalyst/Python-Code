## What should we clean up?

```python
df = bios.copy()
```

```python
len(df)
```

```python
len(df[df["Used name"].str.contains("•")])
```

```python
# Clean the "Used name" column and created the new "Name" column
```

```python
df["Name"] = df["Used name"].str.replace("•", " ")
```

```python
df.head()
```

### Split the Measurements column into height  & weight columns

```python
df[["height_cm", "weight_kg"]] = df["Measurements"].str.split("/", expand=True)
```

```python
len(df[df["Measurements"].str.contains("/", na=False)])
```

```python
len(df.dropna(subset = ["Measurements"]))
```

```python
# Check to make sure all values have the "/"
# get rid of the " cm" & " kg" from our new columns
df[~df["Measurements"].str.contains("/", na=False) & df["Measurements"].notna()]
```

```python
# Making a toy example dataset for "Measurements" column to handle problems
small_data = pd.DataFrame(["183 cm", "180 cm / 150 kg", "190 cm / 80 kg", "100 kg"], columns = ["Measurements"])
small_data[["height_cm", "weight_kg"]] = small_data["Measurements"].str.split("/", expand=True)

small_data["weight_kg"].fillna(small_data["height_cm"], inplace = True)

small_data.head()
```

```python
small_data["height_cm"] = pd.to_numeric(small_data["height_cm"].str.strip(" cm"), errors="coerce")
small_data["weight_kg"] = pd.to_numeric(small_data["weight_kg"].str.strip(" kg"), errors="coerce")
small_data.head()
```

```python
# Now applying on original dataset
df[["height_cm", "weight_kg"]] = df["Measurements"].str.split("/", expand=True)
df["weight_kg"].fillna(df["height_cm"], inplace = True)

df["height_cm"] = pd.to_numeric(df["height_cm"].str.strip(" cm"), errors="coerce")
df["weight_kg"] = pd.to_numeric(df["weight_kg"].str.strip(" kg"), errors="coerce")

df.head()
```

```python
# Finding out whose the tallest man in the olympics and also weighted man
df.sort_values(["height_cm"], ascending = False)
```

```python
# Most Weighted man in the olympics
df.sort_values(["weight_kg"], ascending = False)
```

###  Parse out date from "Born" & "Died" column

```python
date_pattern = r"(\d+\s+\w+\s+\d{4}|\d{4})"
df["born_date"] = df["Born"].str.extract(date_pattern)
```

```python
df.head()
```

### Creating Toy dataset to check "Born" column missing format

```python
small_data = pd.DataFrame(["1895 in scotland", "12 December 1900 in Pakistan", "1977", "23 November 1998 in Santiago, Santiago Metropo..."], columns = ["Born"])
small_data.head()
```

```python
date_pattern = r"(\d+\s+\w+\s+\d{4}|\d{4})"
small_data["born_date"] = small_data["Born"].str.extract(date_pattern)
small_data.head()
```

```python
date_pattern = r"(\d+\s+\w+\s+\d{4}|\d{4})"
df["born_date"] = df["Born"].str.extract(date_pattern)
df["born_year"] = df["Born"].str.extract(r"(\d{4})")

df["born_date"] = pd.to_datetime(df["born_date"], format = "mixed", errors = "coerce")
df["born_year"] = pd.to_numeric(df["born_year"])

df["died_date"] = df["Died"].str.extract(date_pattern)
df["died_date"] = pd.to_datetime(df["died_date"], format = "mixed", errors = "coerce")

df.head()
```

### Get City, Region, Country From "Born" Column

```python
location_pattern = r"in ([\w\s-]+), ([\w\s-]+) \((\w+)\)"
df[["born_city", "born_region", "born_country"]] = df["Born"].str.extract(location_pattern, expand=True)
```
