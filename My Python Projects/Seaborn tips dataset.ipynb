```python
#import the seaborn package
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print(sns.get_dataset_names())
```

```python
#load dataset
df = sns.load_dataset("tips")
df.head()
```

```python
df.shape
```

```python
df.time.unique()
```

```python
df.day.unique()
```

```python
# lineplot seaborn
sns.lineplot(data=df, x="total_bill", y="tip")
plt.title("Seaborn lineplot")
plt.show()
```

```python
# relplot seaborn
sns.relplot(df, x="total_bill", y="tip", kind="scatter")  #relative plot
plt.title("Scatter plot seaborn")
plt.show()
```

```python
sns.relplot(df, x="total_bill", y="tip", kind="line")  #relative plot
plt.title("Line plot seaborn")
plt.show()
```

```python
#Scatter plot using relplot
sns.relplot(df, x="total_bill", y="tip", kind="scatter", col="time",
           hue="smoker", style="smoker", size="smoker")  #relative plot
plt.show()
```

```python
#line plot using line plot
#apply the default theme
sns.set_theme()
sns.relplot(df, x="total_bill", y="tip", kind="line", col="time",
           hue="smoker", style="smoker", size="smoker")  #relative plot
plt.show()
```

```python
#lmplot seaborn
sns.lmplot(data=df, x="total_bill", y="tip", col="time", hue="smoker")
plt.show()
```

```python
#displot seaborn
sns.displot(data=df, x="total_bill", col="time", kde=True)
plt.show()
```

```python
#displot seaborn empirical commulative distribution function 
sns.displot(data=df, x="total_bill", col="time", 
           kind="ecdf", hue="smoker", rug="True",
           rug_kws={"color":"r", "alpha":0.4, "linewidth":2, "height":0.2})
plt.show()
```

```python
#catplot using kind ="swarm"
sns.catplot(data=df, x="day", y="total_bill", kind="swarm", hue="smoker")
plt.show()
```

```python
#catplot using kind ="violin"
sns.catplot(data=df, x="day", y="total_bill", kind="violin", hue="smoker",
           split="True")
plt.show()
```

```python
#catplot using kind ="bar"
sns.catplot(data=df, x="day", y="total_bill", kind="bar", hue="smoker")
plt.show()
```

```python
#Boxplot
sns.boxplot(data=df, x=df["total_bill"])
plt.show()
```

```python
#countplot
sns.countplot(data=df, x="sex")
plt.show()
```

```python
#stripplot
sns.stripplot(data=df, x="total_bill", y="day", hue="day", legend=False)
plt.show()
```

```python
#stripplot
sns.stripplot(data=df, x="total_bill", y="day", hue="sex")
plt.show()
```

```python
#stripplot
sns.stripplot(data=df, x="total_bill", y="day", hue="size")
plt.show()
```

```python
#stripplot
sns.stripplot(data=df, x="total_bill", y="day", hue="size", palette="deep", dodge=True, jitter=False)
plt.show()
```

```python
#stripplot
sns.stripplot(data=df, orient="h")
plt.show()
```

```python
#boxenplot
sns.boxenplot(x=df["total_bill"])
plt.show()
```

```python
sns.boxenplot(data=df, x="day", y="total_bill", k_depth="trustworthy", trust_alpha=0.02)
plt.show()
```
