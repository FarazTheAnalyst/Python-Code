```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.style.use("fivethirtyeight")
```

```python
data = pd.read_csv(r"C:\Users\Dell\Downloads\salaries.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
ages = data["Age"]
dev_salaries = data["All_Devs"]
py_salaries = data["Python"]
js_salaries = data["JavaScript"]
```

### Two Axis

```python
plt.figure(figsize=(8, 3))
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)

ax1.plot(ages, dev_salaries, label="All Dev",
        color="#444444",
        linestyle = "--", 
        )

ax2.plot(ages, py_salaries, label="Python")
ax2.plot(ages, js_salaries, label="JavaScript")

ax1.legend(fontsize=10)
ax1.set_title("Median Salary (USD) by Age", fontsize=16)
ax1.set_ylabel("Median Salary (USD)", fontsize=12)

ax2.legend(fontsize=10)
ax2.set_xlabel("Ages", fontsize=12)
ax2.set_ylabel("Median Salary (USD)", fontsize=12)

plt.tight_layout()

plt.show();
```
