### Load Nessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

#plt.plot(x, y, color="#ababab", linewidth=2, marker=".", linestyle="--", markersize=10, markeredgecolor='blue')

plt.figure(figsize=(5,3), dpi=100)
#Use shorthand notation
#fmt=[color][marker][line]
plt.plot(x, y, "g^--", label="2x")

#line number two
x2=np.arange(0, 4.5, 0.5)
plt.plot(x2[:5], x2[:5]**2, "r", label="x^2")
plt.plot(x2[4:], x2[4:]**2, "r--")

plt.title("our first plot", fontdict={"fontname" : "Comic Sans MS", "fontsize":20})
plt.xlabel("x axis", fontdict={"fontname": "Comic Sans MS"})
plt.ylabel("y axis")

plt.xticks([0, 1, 2, 3])
#plt.yticks([0, 2, 4, 6, 7, 7.5, 8, 10])

plt.legend()
plt.savefig("mygraph.png", dpi=300)
plt.show()
```

### Bar Plot

```python
labels=["A", "B", "C"]
values=[1,4,2]

bars=plt.bar(labels, values)

patterns=["/", "o", "*"]
for bar in bars:
    bar.set_hatch(patterns.pop(0))
    
# bars[0].set_hatch("/")
# bars[1].set_hatch("o")
# bars[2].set_hatch("*")
```
