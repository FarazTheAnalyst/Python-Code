```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns

cars = sns.load_dataset("mpg")

cars.head()
```

```python
plt.scatter(cars.weight, cars.mpg, s=70, c=cars.cylinders, cmap="gray")

plt.legend()
```

```python
plt.scatter(cars.weight, cars.mpg, s=70, c=cars.cylinders, cmap="Accent")
```

```python
plt.scatter(cars.weight, cars.mpg, s=70, c=cars.cylinders, cmap="Accent_r")
```

#### Color scatterplot by Category

```python
groups = cars.groupby("origin")


for name, group in groups:
    plt.plot(group.weight, group.mpg, marker="o", markersize=5, linestyle="", alpha=0.5, label=name )

plt.xlabel('Weight (lbs)')
plt.ylabel("Miles per (Gallon)")
plt.title("Weight vs Mpg")
    
plt.legend()
```
