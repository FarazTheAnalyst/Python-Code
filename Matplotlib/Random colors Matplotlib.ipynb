#### Genereate random colors for Line plot

```python
import matplotlib.pyplot as plt
import numpy as np

#define data
x = [1, 2, 3, 4, 5]
y = [7, 12, 15, 19, 25]

# define random colors
col = (np.random.random(), np.random.random(), np.random.random())

#create lineplot with random color
plt.plot(x, y, c= col)
```

#### Generate ranodm colors for Scatterplot

```python
#define data
x = [1, 2, 3, 4, 5]
y = [7, 12, 15, 19, 25]

#create scatterplot with random colors for each point
plt.scatter(x=x, y=y, c=np.random.rand(len(x), 3))
```
