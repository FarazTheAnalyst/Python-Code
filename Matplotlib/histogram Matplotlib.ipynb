### Part2: using csv file

```python
#real world stackoverflow dataset
data = pd.read_csv(r"C:\Users\Dell\Downloads\age_sal.csv")

data.head()
```

```python
data.shape
```

```python
ids = data["Responder_id"]
ages = data["Age"]

bins = [10, 20, 30, 40, 50, 60]
```

```python
st.median(ages)
```

```python
plt.hist(ages, bins=bins,
         edgecolor="black",
        log=True)

median_age = 29
color = "#fc4f39"

plt.axvline(median_age, color=color,
            label="Median_age",
           linewidth=2
           )

plt.legend()

plt.title("Age of Repondents")
plt.xlabel("Ages")
plt.ylabel("Total Respondents")

plt.tight_layout()
```
