### scatterplot: Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use("seaborn")
```

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
```

### scatterplot

```python
plt.scatter(x, y,
           c="green",
            s=100
           )
```

```python
plt.scatter(x, y,
           c="green",
           marker="X",
            s=100
           )
```

```python
 colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
    
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,
         538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]


plt.scatter(x, y,
           c=colors,
            cmap="Greens",
            edgecolor="black",
            linewidth=1,
            alpha=0.5,
            s=sizes
           )

cbar = plt.colorbar()
cbar.set_label("Satisfiction")
```

### Part 2: Using  real wourld csv file

```python
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\Dell\Downloads\youtube_data.csv")

data.shape
```

```python
data.head()
```

```python
view_count = data["view_count"]
likes = data["likes"]
ratio = data["ratio"]

plt.scatter(view_count, likes,
            edgecolor="black",
            linewidth=1,
            alpha=0.5)
```

```python
plt.scatter(view_count, likes,
            c=ratio,
            cmap="summer",
            edgecolor="black",
            linewidth=1,
            alpha=0.75
           )

cbar = plt.colorbar()
cbar.set_label("Likes/Dislikes ratio")



plt.xscale("log")
plt.yscale("log")

plt.title("Trending Youtube Videos")
plt.xlabel("View Count")
plt.ylabel("Total Likes")

plt.tight_layout()

plt.show()
```
