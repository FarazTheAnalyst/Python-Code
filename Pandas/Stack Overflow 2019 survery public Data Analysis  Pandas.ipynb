#### Reading Data from sqlserver with query

```python
import pandas as pd
from sqlalchemy import create_engine

# Create the SQLAlchemy engine
driver = 'ODBC Driver 17 for SQL Server'
server = 'DESKTOP-VDP7PIC\\SQLEXPRESS'
database = 'sample_db'

# Ensure the connection string is correctly formatted
connection_string = f'mssql+pyodbc://@{server}/{database}?driver={driver}&trusted_connection=yes'
engine = create_engine(connection_string)

# SQL query to read the first 20 rows
query = 'SELECT TOP 20 * FROM saudi_dev'

# Use pandas to execute the query and read the data into a DataFrame
pak_df_first_20 = pd.read_sql_query(query, engine)

# Display the DataFrame
pak_df_first_20
```
