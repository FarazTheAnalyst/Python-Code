### Working with Dates and Time Series Data

```python
import pandas as pd
```

```python
#Etherium Crypto Currency dataset broken on 1 hour segment
df = pd.read_csv(r"C:\Users\Dell\Downloads\ETH_1h.csv")
df.shape
```

```python
df.head()
```

```python
df.loc[0, "Date"]
```

```python
df["Date"] = pd.to_datetime(df["Date"], format="%Y-%m-%d %I-%p")
```

```python
df.loc[0, "Date"].day_name()
```

```python
df.loc[0:4,"Date"].dt.day_name()
```

```python
df["DayofWeek"] = df["Date"].dt.day_name()
df
```

```python
df["Date"].min()
```

```python
df["Date"].max()
```

```python
min_date = df["Date"].min()
filt = df["Date"]==min_date
df.loc[filt]
```

```python
max_date = df["Date"].max()
filt = df["Date"]==max_date
df.loc[filt]
```

#### TimeDelta: it calulates the difference between to dates that is timedelta

```python
df["Date"].max() - df["Date"].min()
```

```python
filt = (df["Date"] >= "2019") & (df["Date"] < "2020")
df.loc[filt]
```

```python
filt = (df["Date"] >= pd.to_datetime("2019-1-1")) & (df["Date"] < pd.to_datetime("2020-1-1"))
df.loc[filt]
```

#### High value for a Day

```python
df.loc["2020-1-1"]["High"].max()
```

```python
highs = df["High"].resample("D").max()
highs
```

```python
highs_d = df.loc[:,"High"].resample("D").max()
highs_d["2020-1-1"]
```

#### High for each week

```python
highs_w = df.loc["2020-01-01":"2020-1-5"]["High"].max()
highs_w
```

```python
highs_w = df["High"].resample("W").max()
highs_w["2020-01-01":"2020-1-5"]  #we can  also use .lo[]
```
