#### How to Write and Match Regular Expressions (Regex)

```python
import re
```

```python
text_to_search = '''
abcdefghijklmnopqurtuvwxyz
ABCDEFGHIJKLMNOPQRSTUVWXYZ
1234567890

Ha HaHa

MetaCharacters (Need to be escaped):
. ^ $ * + ? { } [ ] \ | ( )

coreyms.com

321-555-4321
123.555.1234
123*555*1234
800-555-1234
900-555-1234

Mr. Schafer
Mr Smith
Ms Davis
Mrs. Robinson
Mr. T
'''

sentence = 'Start a sentence and then bring it to an end'
```

```python
pattern = re.compile(r"abc")
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# if we use (.) with out excape \
pattern = re.compile(r".")
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match) #it returns weird all characters
```

```python
#so we need to use escape \ with all the MetaCharacter (Need to be excaped) 
# . ^ $ * + ? {} [] \ | ()
pattern = re.compile(r"\.")
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
pattern = re.compile(r"coreyms\.com")
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
"""
.       - Any Character Except New Line
\d      - Digit (0-9)
\D      - Not a Digit (0-9)
\w      - Word Character (a-z, A-Z, 0-9, _)
\W      - Not a Word Character
\s      - Whitespace (space, tab, newline)
\S      - Not Whitespace (space, tab, newline)

\b      - Word Boundary
\B      - Not a Word Boundary
^       - Beginning of a String
$       - End of a String

[]      - Matches Characters in brackets
[^ ]    - Matches Characters NOT in brackets
|       - Either Or
( )     - Group

Quantifiers:
*       - 0 or More
+       - 1 or More
?       - 0 or One
{3}     - Exact Number
{3,4}   - Range of Numbers (Minimum, Maximum)"""
```

```python

# . - Any Character Except New Line
pattern = re.compile(r".")
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# \d      - Digit (0-9)
pattern = re.compile(r"\d")
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# \D    -Not a digit (0-9)
pattern = re.compile(r"\D")
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# \w   -Word Character (a-z, A-Z, 0-9, _)
pattern = re.compile(r"\w")
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
data=[]
for i in range(1, 51):
    name = f"Faraz Gill {i}"
    data.append([name])
    
for i in data:
    i[0] = re.sub(r"\d", "", i[0])
    
print(data)
```

```python
# \W    - Not a Word Character
pattern = re.compile(r"\W")
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# \s    - Whitespace (space, tab, newline)
pattern = re.compile(r"\s")
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# \S   - Not a Whitespace (space, tab, newline)
pattern = re.compile(r"\S")
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

#### Anchors

```python

# \b   - Word Boundary
pattern = re.compile(r"\bHa")
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# \B   - Not a Word Boundary
pattern = re.compile(r"\BHa")
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# ^   - Beggining of a String
pattern = re.compile(r"^Start")
matches = pattern.finditer(sentence)

for match in matches:
    print(match)
```

```python
# $   - End of a String
pattern = re.compile(r"end$")
matches = pattern.finditer(sentence)

for match in matches:
    print(match)
```
