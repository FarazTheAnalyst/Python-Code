```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cars = sns.load_dataset("mpg").dropna()

cars.shape
```

```python
cars.head()
```

### Univariate (1D)

```python
sns.kdeplot(cars.horsepower, shade=True);
```

```python
sns.kdeplot(cars.horsepower, bw=10);
```

```python
sns.kdeplot(cars.horsepower, cumulative=True)
```

### Bivariate Kde (2d)

```python
sns.kdeplot(x=cars.horsepower, y=cars.mpg, n_levels=20)
plt.show()
```

```python
sns.kdeplot(x=cars.horsepower, y=cars.mpg, 
            shade=True);
```

```python
sns.kdeplot(x=cars.horsepower, y=cars.mpg, 
           shade=True,
           cbar=True);
```

```python
sns.set_style('whitegrid')

cyl_4 = cars[cars.cylinders == 4]
cyl_8 = cars[cars.cylinders == 8]

plt.figure(figsize=(8, 6))

sns.kdeplot(x = cyl_4.horsepower, y = cyl_4.mpg,
            cmap="Blues", shade=True)
sns.kdeplot(x = cyl_8.horsepower, y = cyl_8.mpg,
            cmap="Reds", shade=True)

plt.xlabel('Horsepower', fontsize=14)
plt.ylabel('Miles per Gallon (MPG)', fontsize=14)

# Correct the color variables to be strings
plt.annotate("4 Cylinders", (105, 32), color='blue', fontsize=16, fontweight='bold')
plt.annotate("8 Cylinders", (190, 18), color='red', fontsize=16, fontweight='bold')

plt.show()
```

### Kde with hist

```python
plt.figure(figsize=(8,5))
sns.distplot(cars.horsepower, kde_kws={"lw":5}, hist_kws={"alpha":0.25})

sns.despine(left=True)
plt.xlim(0, 280)
plt.xlabel("")
plt.ylabel("")
plt.yticks([])
plt.tight_layout();
```

### Pairplot

```python
sns.pairplot(data = cars[cars["cylinders"].isin([4, 6, 8])],
            hue="cylinders",
            vars = ["mpg", "horsepower", "acceleration"],
            palette = "inferno_r"
            );
```
