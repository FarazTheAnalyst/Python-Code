### Highliting Palette

```python
palette_dict = {x: "gray" for x in tips.day.unique()}
```

```python
palette_dict
```

```python
palette_dict["Sat"] = "crimson"
```

```python
palette_dict
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="day", palette=palette_dict)
```

```python
sns.displot(x="tip", data=tips, hue="day", col="day", palette=palette_dict)
```

## Part Two: Advance  Palettes and Palletes Widgets

### Advance Palettes

#### This system creates a palette with linearly_increasing or -decreasing brightness, which means information can be preserved for both:

* black-and-white printer
* individual experiencing colorblindness

```python
sns.cubehelix_palette(start=1, rot=1, light=0.75,dark=0.4,n_colors=10)
```

#### A short hand string can also be passed to color_palette() to achieve the same effect

```python
sns.color_palette("ch:1,r=1,l=0.75,d=0.4",10)
```

### husl_palette()

#### Based of the HUSL color system (now re-named HSluv), a human friendly version of HSL color system.
#### Retrieved evenly spaced colors from the HUSL color space,  Arguments include:

* h: first hue, starting color
* s: (0 = gray 1 = more colorful)
* l: lightness(0 = dark, 1 = light)

```python
sns.husl_palette(h=0.5, s=0.5, l=0.4)
```
