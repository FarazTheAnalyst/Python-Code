### scatterplot: Seaborn

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
diamonds = sns.load_dataset("diamonds")

diamonds.shape
```

```python
diamonds.head()
```

```python
diamonds = diamonds[
    diamonds.cut.isin(["Premium", "Good"]) &  # Filter rows where cut is Premium or Good
    diamonds.color.isin(["D", "F", "J"])      # Filter rows where color is D, F, or J
].sample(n=100, random_state=22)             # Sample 100 rows from the resulting DataFrame


diamonds.shape
```

```python
# Check the filtered DataFrame
print(diamonds['cut'].unique())
```

```python
diamonds.dtypes
```

```python
import pandas as pd

diamonds["cut"] = pd.Categorical(diamonds["cut"], categories = ["Premium", "Good"])
diamonds["color"] = pd.Categorical(diamonds["color"], categories = ["D", "J", "F"])
```

```python
# Check the filtered DataFrame
print(diamonds['cut'].unique())
```

```python
diamonds.head()
```

### hue + style + size

```python
sns.scatterplot(x="carat", y="price", data=diamonds,
               hue="cut",
                style="cut",
               s=100)
```

### alpha

```python
sns.scatterplot(x="carat", y="price", 
               data=diamonds,
               alpha=0.6,
                s=100,
                hue="cut",
                style="color"
               );
```

### marker style

```python
sns.scatterplot(x="carat", y="price",
               data=diamonds,
                marker="*",
                s=200
               );
```

```python
sns.scatterplot(x="carat", y="price",
               data=diamonds,
                marker="*",
               edgecolor="black",
                s=200
               );
```
