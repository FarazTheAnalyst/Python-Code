### clustermap: Seaborn

```python
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

### Basics - Toy Example

```python
toy_data_dict = {
    "Name" : ["Andy", "Sen", "Claire", "David"],
    "Hours_Studied" : [10, 20, 12, 28],
    "Score" : [15, 20, 16, 25],
    "Street_Adress" : [11, 24, 18, 9]
}
```

```python
toy_data = pd.DataFrame(toy_data_dict)
toy_data.set_index("Name", inplace=True)

toy_data
```

```python
sns.clustermap(toy_data, figsize=(4, 4), annot=True);
```

### Basics - Penguin Dataset

#### Load the data


```python
penguins = sns.load_dataset("penguins").dropna()
```

```python
penguins.head()
```

```python
penguins.shape
```

### Build clustermap

```python
# numerical_cols = penguins.select_dtypes(include = ["float64", "int64"])
numerical_cols = penguins.columns[2:6]
```

```python
sns.clustermap(penguins[numerical_cols])
```

```python
sns.clustermap(penguins[numerical_cols], standard_scale=1)
```

### Linkage and Metric

#### Method = Linkage

```python
sns.clustermap(penguins[numerical_cols], standard_scale=1, method="single")
```

```python
sns.clustermap(penguins[numerical_cols], standard_scale=1, metric="cityblock")  #mahatten distance
```

### Additioal Options

#### row_colors

```python
species_colors=penguins.species.map({
    "Adelie" : "blue",
    "Chinstrap" : "red",
    "Gentoo" : "green"
})
```

```python
len(species_colors)
```

```python
species_colors[:5]
```

```python
sns.clustermap(penguins[numerical_cols], standard_scale=1, row_colors = species_colors);
```
