### relplot: Seaborn

```python
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
```

```python
df.head()
```

```python
df.shape
```

#### Create new column with age decade bucket for each passenger

```python
df["decade"] = [age // 10*10 for age  in df.age]
```

```python
df.head()
```

### Intro Visual

#### Basics

```python
sns.set_style("white")
```

```python
sns.relplot(x="age", y="fare",
            data=df,
           hue="pclass",
            style="pclass",
            size="pclass"
           );
```

### Kind

```python
sns.relplot(x="decade", y="fare",
           data=df,
           kind="line",
            hue="pclass",
            style="pclass"
           );
```

### Small Multipules (FacetGrid)
### row, col

```python
sns.set(style="white", font_scale=1.5)

sns.relplot(x="age", y="fare",
            data=df,
            col="pclass",
           hue="pclass"
           )
```

```python
sns.relplot(x="decade", y="fare",
           data=df,
           kind="line",
           col="pclass",
            hue="pclass"
           );
```

```python
sns.relplot(x="decade", y="fare",
           data=df,
           kind="line",
           row="pclass",
           hue="pclass"
           );
```

### col_wrap

```python
sns.relplot(x="age", y="fare",
           data=df,
           kind="scatter",
           col="deck",
           col_wrap=3
           );
```

### Styling

#### In general, additional keyword arguments will be passed to the kind  of a plot selected

```python
sns.relplot(x="age", y="fare",
           data=df,
           col="pclass",
           hue="survived",
            s=150,
            palette=["black", "red"]
           );
```

### Marker and sizes

```python
sns.relplot(x="age", y="fare", 
          data=df,
            col="pclass",
#           s=150,
          hue="survived",
          style="survived",
            size="survived",
           palette=["black", "red"],
            markers=["X", "o"],
            sizes=(130, 80)
           )
```

### FacetGrid Styling

```python
g = sns.relplot(x="decade", y="survived",
           data=df[df.age < 70],
           kind="line",
            hue="sex",
            errorbar=None,
                col="pclass"
           );

g.set_xlabels("Passenger Age Group")
g.set_ylabels("Percent Survived")
```
