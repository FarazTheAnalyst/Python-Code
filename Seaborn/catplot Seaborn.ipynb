### catplot: Seaborn

```python
import seaborn as sns
```

```python
#load car data and drop rows with missings
cars = sns.load_dataset("mpg").dropna()

#filters to 4-, 6-, 8-, cylinder cars
cars = cars[cars.cylinders.isin([4, 6, 8])]

#add old/new column
cars["type"] = ["old" if x <= 76 else "new" for x in cars.model_year]
```

```python
cars.head()
```

```python
cars.shape
```

### Intro Visuals

#### Basics

```python
sns.set_style("white")
```

```python
sns.catplot(y="horsepower", data=cars);
```

### Kind

```python
sns.catplot(x="horsepower", data=cars,
           kind="box")
```

```python
sns.catplot(x="horsepower", data=cars, 
           kind="violin")
```

### Categorical Varibles, hue

```python
sns.catplot(x="origin", y="horsepower",
            data=cars,
           kind="box",
            hue="cylinders"
           );
```

### Small Multiples (FacetGrid)

```python
sns.set_style("darkgrid")
```

#### row, col

```python
sns.catplot(x="cylinders", y="horsepower", 
           data=cars,
           kind="strip",
            col="origin",
            row="type",
            hue="cylinders"
           );
```

#### row_order (order, hue_order, col_order)

```python
sns.catplot(x="cylinders", y="horsepower", 
           data=cars,
           kind="strip", 
           col="origin",
           row="type", 
           hue="cylinders", 
           row_order=["new", "old"],
            col_order=["usa", "europe", "japan"],
            order=[8, 6, 4]
           );
```

### Styling
##### in general, additional keyword arguments will be passed to the kind of plot selected

```python
sns.catplot(x="cylinders", y="horsepower",
           data=cars,
           kind="box",
           col="origin",
           palette="winter",
           whis=1
           );
```

```python
sns.catplot(x="cylinders", y="horsepower",
           data=cars,
           kind="strip",
           col="origin", 
           size=12,
           alpha=0.3,
           linewidth=2,
            hue="cylinders",
            palette="autumn"
           );
```

### FacetGrid Styling

```python
g = sns.catplot(x="cylinders", y="horsepower",
               data=cars,
               col="origin",
               kind="box"
               );
```
