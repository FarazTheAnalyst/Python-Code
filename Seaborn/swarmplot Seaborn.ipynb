### swarmplot: Seaborn

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset("mpg").dropna()

cars.shape
```

```python
cars.head()
```

### intro visuals

#### Basics

```python
cars.cylinders.value_counts()
```

### Filter down only cars with even number of cylinders

```python
cars = cars[cars.cylinders.isin([4, 6, 8])]
```

```python
sns.swarmplot(x=cars.displacement)
```

```python
cars.displacement.value_counts()
```

```python
sns.swarmplot(x=cars.origin, y=cars.horsepower,
              size=4
             );
```

```python
sns.swarmplot(x="origin", y="horsepower", data=cars,
              size=4
             );
```

```python
sns.swarmplot(x="origin", y="horsepower", data=cars,
             hue="cylinders",
              dodge=True,
              size=4.5
             );
```

### swarmplot Overlay

```python
usa = cars[cars["origin"]=="usa"]

usa.head()
```

```python
import numpy as np

sns.boxplot(x=usa.cylinders, y=usa.horsepower,
           whis=np.inf)
sns.swarmplot(x=usa.cylinders, y=usa.horsepower,
             size=4.8,
              color="black",
              alpha=0.5
             );
```

### With violinplot

```python

sns.violinplot(x=usa.cylinders, y=usa.horsepower,
              scale="width",
              inner=None,
              color="xkcd:forest green")

sns.swarmplot(x=usa.cylinders, y=usa.horsepower,
              size=3.5,
             color="white"
             )

plt.ylim(46, 255);
```

### Styling

#### linewidth, edgecolor

```python
sns.swarmplot(cars.horsepower,
             linewidth=1,
              edgecolor="black"
             );
```

### Matplotlib Scatter Properties: marker

```python
plt.figure(figsize=(8, 4))

sns.swarmplot(x=cars.horsepower, marker="*",
             linewidth=1,
             edgecolor="black")
```
