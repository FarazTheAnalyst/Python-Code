### displot: Seaborn

```python
import  seaborn as sns
```

```python
cars = sns.load_dataset("mpg").dropna()

cars.head()
```

### Basics

```python
sns.set_style("white")
```

```python
sns.displot(x="weight", data=cars, kde=True)
```

### Kind - *NEW!*

#### Kernal Density estimation Plot: kind="kde"

```python
sns.displot(x="weight", data=cars,
           kind="kde",
           linewidth=4)
```

### Empirical Cumulative Distirbutioin Function Plot: Kind = "ecdf"

```python
sns.displot(x="weight", data=cars,
           kind="ecdf")
```

#### rug

```python
import matplotlib.pyplot as plt

sns.displot(
    x="weight", 
    data=cars,
    kind="kde", 
    rug=True,
    rug_kws={"height": 0.05, "linewidth": 0.2},  # Adjust the height of the rug plot
);
```

### Categorical Varibles: hue

```python
sns.displot(x="weight", data=cars,
            hue="origin"
           )
```

```python
sns.displot(x="weight", data=cars,
           kind="kde",
            hue="origin"
           );
```

### Bivariate (2D) Options - NEW!

```python
sns.displot(x="weight", y="horsepower", 
            data=cars
           );
```

```python
sns.displot(x="weight", y="horsepower", 
           data=cars,
           kind="kde",
            rug=True,
            rug_kws={"height": 0.05, "linewidth":0.2}
           );
```

### Small Mulitples (FacetGride) - NEW!

```python
sns.displot(x="weight", data=cars,
           hue="origin",
            col="origin"
           );
```

```python
sns.set(style="white", font_scale=1.5)

sns.displot(x="weight", data=cars,
           kind="kde",
           hue="origin", 
           col="origin",
            fill=True,
            rug=True,
            rug_kws={"height":0.05, "linewidth":0.2}
           );
```

```python
g = sns.displot(x="weight", data=cars[cars.cylinders.isin([4, 6, 8])],
            col="origin",
           row="cylinders",
           hue="cylinders"
           );

g.set_titles(col_template = "{col_name} origin")
```

### Styling

#### palette

```python
sns.displot(x="weight", data=cars,
            hue="origin",
           col="origin",
           palette="crest"
           );
```

### height, aspect

```python
sns.displot(x="weight", data=cars,
           col="origin",
           height=5,
            aspect=1.5
           );
```

### Other Keywords: Plot 

#### in general, additional keyword arguments will be passed  to the kind of plot selected

```python
sns.displot(x="weight", data=cars, 
           bins=20
           );
```

```python
sns.displot(x="weight", data=cars,
           kind="kde", 
           fill=True,
           linewidth=4);
```

### rug_kws


```python
sns.displot(x="weight", data=cars,
            kind="kde", 
            rug=True, 
            rug_kws={"height":0.05, "linewidth":0.2}
           );
```

```python
### FacetGrid Styling
#### facet_kws
g = sns.displot(x="weight", data=cars,
           col="origin",
           facet_kws={"despine":False})
```

```python
type(g)
```

```python
g.axes_dict.items()
```

```python
g = sns.displot(x="weight", data=cars,
             col="origin"
             );

for col_val, ax in g.axes_dict.items():
    if col_val == "europe":
        ax.set_facecolor("#E2F2FF")
    
```

### Example Thumbnails

```python
from matplotlib import pyplot as plt

plt.rcParams["font.size"] = 14
```

```python
sns.displot(x="weight", y="horsepower", data=cars)
plt.xlabel("Weight", fontsize=16)
plt.ylabel("Height", fontsize=16);
```

```python
g = sns.displot(x='weight', row='origin', hue='origin', palette='crest', 
            data=cars, height=4, aspect=2)
g.set_titles(row_template="{row_name} cars")
plt.xlabel("Weight")
plt.xticks([2000, 3000, 4000, 5000]);
```

```python
g = sns.displot(x="weight",
               color="#8711ac",
               data=cars,
               height=4, aspect=0.8,
               kde=True)
g.set_titles(col_template="")
g.set_xlabels("")
g.set_ylabels("")
g.set_xticklabels([])
g.set_yticklabels([])
plt.xlim(1500, 5500)
plt.ylim(0, 80)
plt.tight_layout();
```

```python
g = sns.displot(x="weight", y="horsepower", row="origin", color="#48804D",
               data=cars[cars.origin.isin(["europe", "japan"])],
               height=3, aspect=2.3, row_order=["europe", "japan"],
               facet_kws={"despine":False})
g.set_titles(row_template="")
g.set_xlabels("")
g.set_ylabels("")
g.set_xticklabels([])
g.set_yticklabels([])
plt.tight_layout();
```
