### lmplot: Seaborn

```python
import seaborn as sns
```

```python
df = sns.load_dataset("penguins").dropna()

df.head()
```

### Basics

```python
sns.set_style("white")
```

```python
sns.lmplot(x="bill_length_mm", y="bill_depth_mm"
           data=df
          );
```

```python
sns.lmplot(x="bill_length_mm", y="bill_depth_mm",
          data=df,
          hue="species"
          );
```

### markers

```python
sns.lmplot(x="bill_length_mm", y="bill_depth_mm", 
           data=df,
          hue="species",
          markers=["o", "x", "d"]
          );
```

### Small Mulitples (FacetGrid)

#### col

```python
sns.lmplot(x="bill_length_mm", y="bill_depth_mm", 
          data=df,
          hue="species",
          col="species"
          );
```

#### row

```python
sns.set(style="white", font_scale=1.5)

sns.lmplot(x="bill_length_mm", y="bill_depth_mm",
          data=df,
          hue="species",
           col="species",
           truncate=False,
          row="sex"
          );
```

### Extras

#### x_bins

```python
sns.lmplot(x="bill_length_mm", y="bill_depth_mm",
          data=df[df.species=="Chinstrap"],
           x_bins=5
          );
```

### Other models: Logistic Regression, Logistic

```python
df["sex_binary"] = [(x == "Male") for x in df.sex]
```

```python
df.head()
```

```python
sns.lmplot(x="body_mass_g", y="sex_binary", 
          data=df,
           logistic=True,
           n_boot=100,
          hue="species",
           col="species"
          )
```

### Styling

#### scatter_kws

```python
sns.lmplot(x="bill_length_mm", y="bill_depth_mm",
          data=df,
          hue="species",
          col="species",
           scatter_kws={"s":150, "edgecolor":"black"}
          )
```

### line_kws

```python
sns.lmplot(x="bill_length_mm", y="bill_depth_mm",
          data=df,
          hue="species",
          col="species",
          line_kws={"lw" : 5, "ls" : "-", "color":"black"})
```

#### FacetGrid Object

```python
g = sns.lmplot(x="bill_length_mm", y="bill_depth_mm",
              data=df,
              col="species",
              hue="species"
              )
g.set_titles(col_template="{col_name} penguins")
g.set_xlabels("Bill Depth (mm)");
```

### Examples, Thumbnail

```python
from matplotlib import pyplot as plt
```

```python
sns.set(style="white", font_scale=1.7)
g = sns.lmplot(x="bill_length_mm", y="bill_depth_mm",
               data=df,
               col="species",
               hue="species",
               truncate=False
              )
g.set_titles("{col_name} Penguins")
g.set_xlabels("Bill Length (mm)")
g.set_ylabels("Bill Depth (mm)")
plt.tight_layout()
#plt.savefig("20Lmplot_image_facetGridExample.png", dpi=200);
```

```python
diamonds = sns.load_dataset("diamonds").sample(1000)
```

```python
sns.lmplot(x="carat", y="price", data=diamonds, hue="clarity", legend=False)
plt.xlabel("Carat")
plt.ylabel("Price")
plt.ylim(0, None)
plt.xlim(0, None)
plt.legend(loc=2, title="clarity")
plt.tight_layout()
#plt.savefig("20Lmplot_image_tooManyLines.png", dpi=200);
```

```python
sns.set(style="white", font_scale=2)
sns.lmplot(x="carat", y="price", hue="clarity", data=diamonds, col="clarity", col_wrap=3)
plt.xlim(0,None)
plt.ylim(0, None)
plt.savefig("20Lmplot_image_tooManyLinesFacet.png", dpi=200);
```

### Thumb

```python
sns.set(style="white", font_scale=1)
g = sns.lmplot(x="bill_length_mm", y="bill_depth_mm",
              data=df,
               hue="species",
              col="species",
              truncate=False,
              aspect=0.6,
              palette=["#cb416b", "#001146", "#75bbfd"]
              )

g.set_titles("")
g.set_xlabels("")
g.set_ylabels("")
plt.xticks([])
plt.yticks([])
plt.tight_layout()
#plt.savefig("20Lmplot_image_thumb.png", dpi=200);
```
