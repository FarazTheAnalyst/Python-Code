### Basics

```python
sns.set_style("whitegrid")
```

```python
cars.cylinders.value_counts()
```

#### Filter down only cars with even number of cylinders

```python
cars = cars[cars.cylinders.isin([4, 6, 8])]
```

```python
sns.boxplot(cars.mpg)
```

```python
cars.mpg.describe()
```

```python
sns.boxplot(x=cars.origin, y=cars.mpg);
```

```python
sns.boxplot(x="origin", y="mpg", data=cars);
```

### hue
#### by cylinders

```python
sns.boxplot(x="origin", y="mpg", hue="cylinders", data=cars);
```

### By Model Year

```python
cars.model_year.describe()
```

```python
cars["newer_model"] = cars.model_year > 76
cars.head()
```

```python
sns.boxplot(x=cars.origin, y=cars.mpg, hue="newer_model", data=cars)
```

#### Styling Options

#### Order

```python
sns.boxplot(x="mpg", y="origin",
           data=cars);
```

```python
sns.boxplot(x="mpg", y="origin",
           data=cars,
           order=["japan", "europe", "usa"]);
```

```python
sns.boxplot(x="mpg", y="origin", hue="newer_model",
           data=cars,
           order=["japan", "europe", "usa"],
            hue_order=[True, False]
           );
```

```python
sns.boxplot(x="mpg", y="origin", data=cars,
           hue="newer_model",
           color="g");
```

### width

```python
sns.boxplot(x="mpg", y="origin", data=cars,
           width=0.5
           );
```

### linewidth

```python
sns.boxplot(x="mpg", y="origin", data=cars,
           linewidth=2.5
           );
```

```python
sns.boxplot(x="mpg", y="origin", data=cars,
           whis=1
           );
```

### fliersize

```python
sns.boxplot(x="mpg", y="origin", data=cars,
           fliersize=2
           );
```

```python
sns.boxplot(x="mpg", y="origin", data=cars,
           fliersize=10
           );
```
