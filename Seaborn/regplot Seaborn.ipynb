### regplot: seaborn

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
diamonds = sns.load_dataset("diamonds")

diamonds.shape
```

```python
diamonds.head()
```

```python
diamonds = diamonds.sample(n=200, random_state=44)

diamonds.shape
```

### Intro Visuals

#### Basics

```python
sns.set_style("dark")
```

```python
sns.regplot(x=diamonds.carat, y=diamonds.price)
```

```python
sns.regplot(x="carat", y="price", data=diamonds)
```

```python
sns.regplot(x="carat", y="price", data=diamonds,
           fit_reg=False
           );
```

```python
# plt.gcs().set(xlim(0, 2.6))
sns.regplot(x="carat", y="price", data=diamonds,
           scatter=False
           );
```

### regplot Options

#### Confidense intervals

```python
sns.regplot(x="carat", y="price", data=diamonds,
           ci=None
           );
```

### Discrete Variables

```python
import pandas as pd

cut_categories = ["Fair", "Good", "Very Good", "Premium", "Ideal"]
diamonds["cut"] = pd.Categorical(diamonds["cut"], categories=cut_categories, ordered=True)
```

```python
cut_map = {
    "Fair":1,
    "Good":2,
    "Very Good":3,
    "Premium":4,
    "Ideal":5
}

diamonds["cut_value"] = diamonds.cut.map(cut_map)
```

```python
diamonds["cut_value"] = diamonds["cut_value"].astype(int)
```

```python
diamonds.cut_value.value_counts();
```

```python
# Plot with seaborn
sns.regplot(
    x="cut_value", 
    y="price", 
    data=diamonds, 
    x_jitter=0.1,
    line_kws={"color": "salmon"},
    scatter_kws={"color": "orange"}
)
```

### Estimator Aggregate

```python
import numpy as np
```

```python
sns.regplot(x="cut_value", y="price", 
            data=diamonds,
           x_estimator=np.mean
           
           );
```

### Models

#### Ploynomial Regression (order)

```python
sns.regplot(x="carat", y="price", 
           data=diamonds,
           order=2
           );
```

### Robust Regression

```python
x_example = [0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5]
y_example = [0.1, 0.8, 2.2, 2.7, 3.8, 4.5, 6.2, 6.8, 7.9, 9.4, 30.4]
```

```python
sns.regplot(x=x_example,
           y=y_example,
           ci=None, 
           robust=True
           );
```
