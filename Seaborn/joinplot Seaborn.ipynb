### jointplot: Seaborn

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
geyser = sns.load_dataset("geyser")
```

```python
geyser.head()
```

```python
geyser.shape
```

### Intro Visuals

### Basics

```python
sns.set_style("darkgrid")
```

```python
sns.jointplot(x=geyser.waiting, y=geyser.duration)
```

```python
sns.jointplot(x="waiting", y="duration", data=geyser)
```

### Plot kinds

#### There currently six kinds of jointplot allowed in seaborn

##### 1. scatter (default)
##### 2.kde
##### 3.red
##### 4.hist
##### 5.hex
##### 6.resid

### Scatter (default)

```python
sns.jointplot(x="waiting", y="duration", data=geyser,
             kind="scatter"
             );
```

### kde

```python
sns.jointplot(x="waiting", y="duration", data=geyser,
             kind="kde"
             );
```

### reg

```python
sns.jointplot(x="waiting", y="duration", data=geyser,
             kind="reg"
             );
```

### hist

```python
sns.jointplot(x="waiting", y="duration", data=geyser,
             kind="hist"
             );
```

#### hex

```python
sns.jointplot(x="waiting", y="duration", data=geyser,
             kind="hex"
             );
```

#### resid

```python
sns.jointplot(x="waiting", y="duration", data=geyser, 
             kind="resid"
             );
```

### Returns Grid

```python
g = sns.jointplot(x="waiting", y="duration", data=geyser)
```

```python
type(g)
```

```python
g.plot_joint(sns.kdeplot, color="gray", levels=5)
```

#### hue

```python
sns.jointplot(x="waiting", y="duration", data=geyser);
```

```python
sns.jointplot(x="waiting", y="duration", data=geyser,
             hue="kind",
              kind="kde"
             );
```

### Styling

#### color, palette

```python
sns.jointplot(x="waiting", y="duration", data=geyser,
             color="green"
             );
```

```python
sns.jointplot(x="waiting", y="duration", data=geyser,
             hue="kind"
             );
```

```python
sns.jointplot(x="waiting", y="duration", data=geyser,
             hue="kind",
             palette="autumn"
             );
```

### height, space, ration

```python
sns.jointplot(x="waiting", y="duration", data=geyser,
             height=8,
              space=0.1,
              ratio=5
             );
```

### joint_kws

```python
sns.jointplot(x="waiting", y="duration", data=geyser,
             joint_kws={"marker":5}
             );
```

```python
sns.jointplot(x="waiting", y="duration", data=geyser,
             kind="reg",
              joint_kws={"ci":None, "order":2,
                        "line_kws": {"color": "red"},
                       "scatter_kws":{"color":"purple"}})
```

### marginal_kws

```python
sns.jointplot(x="waiting", y="duration", data=geyser,
              kind="reg",
             marginal_kws={"color":"xkcd:wine red"},
             joint_kws={"ci":None, "order":2,
                        "line_kws": {"color": "red"},
                       "scatter_kws":{"color":"green"}}
             )
```

```python
sns.jointplot(x="waiting", y="duration", data=geyser,
             kind="kde",
             marginal_kws={"lw":4, "fill":True})
```
