### JointGrid: Seaborn

```python
import seaborn as sns
```

```python
cars = sns.load_dataset("mpg")
```

```python
cars.head()
```

### Basics

```python
sns.set_style("darkgrid")
```

```python
g = sns.JointGrid();
```

```python
type(g);
```

```python
g = sns.JointGrid(data=cars, x="weight", y="mpg")
```

### .plot()

```python
g = sns.JointGrid(data=cars, x="weight", y="mpg")
g.plot(sns.scatterplot, sns.histplot)
```

```python
g = sns.JointGrid(data=cars, x="weight", y="mpg")
g.plot(sns.regplot, sns.kdeplot)
```

### .plot_joint(), .marginals()

```python
g = sns.JointGrid(data=cars, x="weight", y="mpg")
g.plot_joint(sns.regplot, color="lightgreen", line_kws={"color":"green"})
g.plot_marginals(sns.histplot, kde=True, alpha=0.5, color="lightgreen", )
```

### JointGrid Axes

```python
g = sns.JointGrid()
x, y = cars.weight, cars.mpg
sns.regplot(x=x, y=y, ax=g.ax_joint, color="lightgreen", line_kws={"color":"green"})
sns.kdeplot(x=x, ax=g.ax_marg_x, color="green")
sns.histplot(y=y, ax=g.ax_marg_y, kde=True, color="lightgreen")
```

### hue

```python
g = sns.JointGrid(data=cars, x="weight", y="mpg", hue="origin")
g.plot(sns.scatterplot, sns.kdeplot)
```

```python
g = sns.JointGrid()
x, y=cars.weight, cars.mpg
sns.regplot(x=x, y=y, ax=g.ax_joint)
sns.kdeplot(x=x, ax=g.ax_marg_x, hue=cars.origin)
sns.histplot(y=y, ax=g.ax_marg_y)
```

### Styling

#### General Stylng

```python
g = sns.JointGrid(data=cars, x="weight", y="mpg")
g.plot(sns.scatterplot, sns.histplot, alpha=0.5, ec="black")
```

### Plot - specific keywords

```python
g = sns.JointGrid(data=cars, x="weight", y="mpg")
g.plot_joint(sns.regplot, line_kws={"color":"black"})
g.plot_marginals(sns.kdeplot, fill=True, color="purple")
```

### JointGrid Properties

```python
g = sns.JointGrid(data=cars, x="weight", y="mpg")
g.plot(sns.scatterplot, sns.histplot)
g.set_axis_labels(xlabel="Weight labels (lbs)", ylabel="Miles per Gallon", fontsize=14)
g.savefig("JointGrid_Example.png");
```

### Thumbnail

```python
sns.set_style("white")

g = sns.JointGrid()
x, y = cars.weight, cars.mpg
sns.regplot(x=x, y=y, ax=g.ax_joint, ci=99,
           scatter_kws={"color":"xkcd:lavender", "alpha":0.7, "ec":"black", "s":50},
           line_kws={"color":"black", "lw":3}
           )
sns.kdeplot(x=x, ax=g.ax_marg_x, lw=3, fill=True, color="xkcd:indigo")
sns.histplot(y=y, ax=g.ax_marg_y, lw=2, color="xkcd:indigo", alpha=0.5, ec="xkcd:indigo")
g.set_axis_labels(xlabel="", ylabel="")

ax = g.ax_joint
ax.set_xticklabels([])
ax.set_yticklabels([])
#g.savefig("25jointgrid_imgae_thumb.png", dpi=200)
```
