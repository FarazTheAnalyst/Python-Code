### PairGrid: Seaborn

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
iris = sns.load_dataset("iris")
```

```python
iris.head()
```

### Basics

```python
sns.set_style("darkgrid")
```

```python
numerical_col = iris.select_dtypes(include=["float64", "int64"])
```

```python
g = sns.PairGrid(numerical_col)
```

```python
type(g)
```

### .map()

```python
g = sns.PairGrid(iris)
g.map(sns.scatterplot);
```

```python
g = sns.PairGrid(numerical_col)
g.map(sns.histplot);
```

### Matplotlib Functions

```python
g = sns.PairGrid(numerical_col)
g.map(plt.scatter);
```

### PairGrid Anatomy

#### .map_diag() and .map_offdiag()

```python
g = sns.PairGrid(numerical_col)
g.map_offdiag(sns.scatterplot)
g.map_diag(sns.histplot);
```

#### .map_upper() and .map_lower()

```python
g = sns.PairGrid(numerical_col)
g.map_lower(sns.scatterplot)
g.map_diag(sns.boxplot)
g.map_upper(sns.kdeplot);
```

### corner

```python
g = sns.PairGrid(numerical_col, corner=True)
g.map_offdiag(sns.scatterplot)
g.map_diag(sns.boxplot);
```

### hue

```python
g = sns.PairGrid(iris, hue="species", diag_sharey=False)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.scatterplot)

g.add_legend();
```

### vars

```python
g = sns.PairGrid(iris, hue="species", diag_sharey=False, vars=["petal_length", "petal_width"])
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.scatterplot)

```

```python
g = sns.PairGrid(iris, hue="species",
                x_vars=["sepal_length", "sepal_width", "petal_width"],
                y_vars=["petal_length"])

g.map(sns.scatterplot)
g.add_legend();
```

### Styling

#### Plot-specific Keywords

```python
g = sns.PairGrid(iris, hue="species", diag_sharey=False)
g.map_diag(sns.kdeplot, fill=True, lw=4)
g.map_offdiag(sns.scatterplot, ec="black", alpha=.5, s=80);
```

### height, aspect

```python
g = sns.PairGrid(iris, hue="species", diag_sharey=False, height=1.5, aspect=2)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.scatterplot);
```

### palette

```python
g = sns.PairGrid(iris, hue="species", diag_sharey=False, palette="cividis")
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.scatterplot);
```

```python
g = sns.PairGrid(iris, hue="species", diag_sharey=False, palette="magma")
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.scatterplot)
g.add_legend();
```
