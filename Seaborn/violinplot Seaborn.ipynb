### violinplot seaborn

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset("mpg").dropna()

cars.shape
```

```python
cars.head()
```

### intro visuals

#### Basics

```python
sns.set_style("whitegrid")
```

```python
cars.cylinders.value_counts()
```

#### Filter down only cars with even number of cylinders

```python
cars = cars[cars.cylinders.isin([4, 6, 8])]
```

```python
cars.cylinders.unique()
```

```python
sns.violinplot(cars.displacement)
```

```python
sns.violinplot(x=cars.cylinders, y=cars.displacement)
```

```python
sns.violinplot(x="cylinders", y="displacement", data=cars)
```

```python
sns.violinplot(x="cylinders", y="displacement", hue="origin",
              data=cars
              );
```

```python
sns.violinplot(x="cylinders", y="displacement", data=cars,
               hue="origin",
              order=[8, 6, 4],
               hue_order=["japan", "europe", "usa"]
              );
```

### violinplot Options

```python
sns.violinplot(x="cylinders", y="displacement", 
               hue="origin",
               data=cars[cars.origin.isin(["japan", "europe"])],
#                split=True,  # Split violins for different hue categories
#                inner="quartile",  # Show quartile lines inside violins
#                palette="Set2"  # Choose a color palette
              )

# Add legend
plt.legend(title="Origin", loc='upper left')

# Show plot
plt.show()
```

```python
sns.violinplot(x="cylinders", y="displacement", hue="origin",
              data=cars[cars.origin.isin(["japan", "europe"])],
              split=True,
               inner="quartiles"
              )

plt.legend(loc=2);
```

### scale

```python
sns.violinplot(x="cylinders", y="displacement", hue="origin",
               data=cars[cars.origin.isin(["japan", "europe"])],
              split=True,
               scale="count"
              );

plt.legend(loc=2)
```

```python
sns.violinplot(x="cylinders", y="displacement", hue="origin",
              data=cars[cars.origin.isin(["japan", "europe"])],
              split=True,
               scale="count",
               scale_hue=False
              );

plt.legend(loc=2)
```

```python
(
    cars[cars.origin.isin(["japan", "europe"])]
    .groupby("cylinders")
    .origin
    .value_counts()
)
```

### Styling

#### Box Options

```python
sns.violinplot(x=cars.origin, y=cars.displacement,
               order=["japan", "europe", "usa"]
              );
```

```python
sns.violinplot(x=cars.origin, y=cars.displacement,
              linewidth=3
              );
```

### KDE Options

```python
sns.violinplot(x=cars.origin, y=cars.displacement,
               bw=4
              );
```

```python
sns.violinplot(x=cars.origin, y=cars.displacement,
               bw=0.2
              );
```
