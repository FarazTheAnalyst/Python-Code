```python
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

```python
tips = sns.load_dataset("tips")

tips.shape
```

```python
tips.head()
```

### Intro Visual

#### Basics

```python
sns.set_style("darkgrid")
```

```python
sns.pairplot(tips)
```

```python
tips.dtypes
```

#### Note that the axes tick labels correspond to the bivariate, relational plots  and NOT the distribution

```python
sns.distplot(tips["size"],
            kde=False,
            hist_kws={"alpha":1})
```

### Plot Kinds
#### diag_kind

```python
sns.pairplot(tips, diag_kind="kde")
```

#### kind

```python
sns.pairplot(tips, kind="reg");
```

### Categorical and Specific Variables

```python
tips["weekend"] = tips.day.isin(["Sat", "Sun"])

tips.head()
```

```python
sns.pairplot(tips, hue="weekend")
```

### vars, x_vars, y_vars

```python
sns.pairplot(tips, hue="weekend", vars=["total_bill", "tip"])
```

```python
sns.pairplot(tips, 
            hue="weekend",
            x_vars=["total_bill", "size"],
            y_vars=["tip"],
            kind="reg",
            height=4
            );
```

### Stylng

```python
del tips["weekend"]
```

```python
from ipywidgets import interactive

def size_widgets(height=2.5, aspect=1):
    sns.pairplot(tips, height=height, aspect=aspect)
```

```python
interactive(size_widgets, height=(1, 3.5, 0.5), aspect=(0.5, 2, 0.25))
```

### palette

```python
sns.pairplot(tips, hue="day", palette="plasma")
```

### diag_kws, plot_kws

```python
sns.pairplot(tips, diag_kws={"color":"gray", "alpha":0.6})
```

```python
sns.pairplot(tips, 
             kind="reg",
            plot_kws={"ci":None, "color":"xkcd:salmon",
                      "scatter_kws":{"color":"gray",
                                    "s":3}})
```
