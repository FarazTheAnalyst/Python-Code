### Seaborn: histplot

```python
import seaborn as sns
```

```python
sns.__version__
```

```python
print(sns.get_dataset_names())
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
penguins.shape
```

```python
penguins.dropna(inplace=True)
```

```python
penguins.shape
```

### Basics

```python
sns.set_style("darkgrid")
```

```python
sns.histplot(x=penguins.bill_length_mm);
```

```python
sns.histplot(y = "bill_length_mm", data=penguins)
```

```python
sns.histplot(x="bill_length_mm", data=penguins,
            kde=True);
```

```python
sns.histplot(x="bill_length_mm", data=penguins,
             bins=[30, 40, 45, 50, 60]);
```

### binwidth and binrange

```python
sns.histplot(x="bill_length_mm", data=penguins,
            binwidth=5, binrange=[30, 60])
```

### density:  AREA of histogram sums to one

```python
sns.histplot(x="bill_length_mm", data=penguins,
            stat="density");
```

### probability: HEIGHT of histogram bars to one

```python
sns.histplot(x="bill_length_mm", data=penguins,
            stat="probability", fill=False, element="step",
            cumulative=True);
```

### Categorical Variables
### hue

```python
sns.histplot(x="bill_length_mm", data=penguins,
            hue="species", 
            element="step");
```

```python
sns.histplot(x="bill_length_mm", data=penguins,
            hue="species", 
            element="poly");
```

```python
sns.histplot(x="bill_length_mm", data=penguins,
            hue="species", 
            multiple="layer");
```

```python
sns.histplot(x="bill_length_mm", data=penguins,
            hue="species", 
            multiple="stack");
```

```python
sns.histplot(x="bill_length_mm", data=penguins,
            hue="species", 
            multiple="fill");
```

### Bivariate Histogram

```python
sns.set_style("white")

sns.histplot(x="bill_length_mm", y="bill_depth_mm", 
            data=penguins,
            cbar=True)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(13,6))
sns.histplot(x="bill_length_mm", y="bill_depth_mm", data=penguins,
            hue="species",
            cbar=True)
```

```python
sns.histplot(x="bill_length_mm", y="species", data=penguins,
            hue="species")
```

### Styling
#### Shrink

```python
sns.histplot(x="species", hue="sex", data=penguins,
            multiple="dodge",
            shrink=0.8);
```

### color, palette

```python
sns.histplot(x="bill_length_mm", data=penguins,
            color="indigo")
```

```python
sns.histplot(x="species", hue="sex", data=penguins,
             multiple="dodge",
             shrink=0.8, palette="bone"
            )
```

### fill

```python
sns.histplot(x="bill_length_mm", data=penguins,
            fill=False);
```
