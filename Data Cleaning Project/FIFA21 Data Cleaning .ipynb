### 11. Contract column

```python
fifa["Contract"].dtype
```

```python
fifa["Contract"].unique()
```

```python
for index, row in fifa.iterrows():
    if "On Loan" in row["Contract"] or "Free" in row["Contract"]:
        print(row["Contract"])
```

```python
def extract_contract_info(contract):
    if contract == "Free" or "On Loan" in contract:
        start_date = np.nan
        end_date = np.nan
        career_length = 0
    else:
        start_date, end_date = contract.split(" ~ ")
        start_year = int(start_date[:4])
        end_year = int(end_date[:4])
        career_length = end_year - start_year
    return start_date, end_date, career_length

#Apply fn to contract column & create a new column
new_cols = ["Contract_Start", "Contract_End", "Contract_Length(years)"]
new_data = fifa["Contract"].apply(lambda x: pd.Series(extract_contract_info(x)))

for i in range(len(new_cols)):
    fifa.insert(loc=fifa.columns.get_loc("Contract")+1+i, column=new_cols[i], value=new_data[i])
```

```python
fifa.sample(3)
```

```python
fifa[["Contract", "Contract_Start", "Contract_End", "Contract_Length(years)"]].sample(5)
```

```python
def categorize_contract_status(contract):
    if contract == "Free":
        return "Free"
    elif "On Loan" in contract:
            return "On Loan"
    else:
        return "contract"

# Add new column to dataframe
fifa.insert(loc=fifa.columns.get_loc("Contract_Length(years)")+1, column="Contract_Status" ,value=fifa["Contract"].apply(categorize_contract_status))
fifa.sample(3)
```

### 13. Height

```python
fifa["Height"].dtype
```

```python
fifa["Height"].unique()
```

```python
def convert_heights(height):
    if "cm" in height:
        return int(height.strip("cm"))
    else:
        feet, inches = height.split("'")
        total_inches = int(feet)*12 + int(inches.strip('"'))
        return round(total_inches * 2.50)
    
# Apply fn to Height column
fifa["Height"] = fifa["Height"].apply(convert_heights)
fifa["Height"].unique()
```

```python
fifa = fifa.rename(columns={"Height":"Height(cm)"})
fifa.sample(3)
```
